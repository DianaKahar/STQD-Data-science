# **A quick glance on Matplotlib**
- a 2D plotting library for Python that produces high-quality static plots

```python
import matplotlib.pyplot as plt

# Create some sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Plot the data
plt.plot(x,y)

# Add labels and title
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Simple Plot'); # use semi column to suppress the output msg
```

# **A quick glance on Seaborn**
- built on top of Matplotlib
- provides a high-level interface for drawing attractive and informative statistical graphics.

***Combination of python function

```python
# Create some sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a DataFrame using the data
data = {'x': x, 'y': y}
df = pd.DataFrame(data)
df
# Create a scatter plot using Seaborn
sns.scatterplot(x = 'x', y = 'y', data = df)

# Add labels and title
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Seaborn Scatter plot');
```

# **Individual plots**

```python
# Histogram - plot 1
# alpha=0.8: sets the transparency of the bars
# an alpha value of 0.8 means the bars are somewhat transparent
#normal distribution -> bell shaped curve
#standard normal
rng = np.random.RandomState(12345)
plt.hist(rng.standard_normal(100), bins= 20, color = "pink", alpha = 0.8); #alpha is for the display of bar, bins is for the number of bars
```

```python
# Scatter plot - plot 2
rng = np.random.RandomState(12345)
plt.scatter(x = np.arange(30), y = np.arange(30)+ 3* rng.standard_normal(30));
```

# LINE PLOT IS SUITABLE FOR TIME-RELATED INFORMATION

-X-AXIS CAN NEVER BE A CATEGORICAL DATA

```python
# Line plot : plot 3
rng = np.random.RandomState(12345)
plt.plot(rng.standard_normal(50).cumsum(), color = 'red', linestyle = 'dashed')

plt.xlabel('Days')
plt.ylabel('y-axis')
plt.title('Line plot');
```

### **Place all the subplots into one panel**

1.   **plt.figure()** - create a new figure
2.   **add_subplot** - to create one or more subplots
3. dpi = dot per inch

```python
# create one or more subplots
# 2 x 2; 4 plots in total
fig = plt.figure(dpi = 300)
ax1 = fig.add_subplot(2, 2, 1) #(2, 2) = subplot of 2 x 2; '1' = 1st plot which is top left -> histogram
ax2 = fig.add_subplot(2, 2, 2) #(2, 2) = subplot of 2 x 2; '2' = 2nd plot which is top right -> scatterplot
ax3 = fig.add_subplot(2, 2, 3) #(2, 2) = subplot of 2 x 2; '3' = 3rd plot which is bottom left -> line plot

# dont need the plt function bcos already define in fig varible
ax1.hist(rng.standard_normal(100), bins= 20, color = "pink", alpha = 0.8)
ax2.scatter(x = np.arange(30), y = np.arange(30)+ 3* rng.standard_normal(30))
ax3.plot(rng.standard_normal(50).cumsum(), color = 'green', linestyle = 'dashed');

# how to save figures
plt.savefig("")
```
