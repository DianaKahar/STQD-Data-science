### **Place all the subplots into one panel**

1.   **plt.figure()** - create a new figure
2.   **add_subplot** - to create one or more subplots
3. dpi = dot per inch

```python
# create one or more subplots
# 2 x 2; 4 plots in total
fig = plt.figure(dpi = 300)
ax1 = fig.add_subplot(2, 2, 1) #(2, 2) = subplot of 2 x 2; '1' = 1st plot which is top left -> histogram
ax2 = fig.add_subplot(2, 2, 2) #(2, 2) = subplot of 2 x 2; '2' = 2nd plot which is top right -> scatterplot
ax3 = fig.add_subplot(2, 2, 3) #(2, 2) = subplot of 2 x 2; '3' = 3rd plot which is bottom left -> line plot

# dont need the plt function bcos already define in fig varible
ax1.hist(rng.standard_normal(100), bins= 20, color = "pink", alpha = 0.8)
ax2.scatter(x = np.arange(30), y = np.arange(30)+ 3* rng.standard_normal(30))
ax3.plot(rng.standard_normal(50).cumsum(), color = 'green', linestyle = 'dashed');

# how to save figures
plt.savefig("") # add name of figures that you want
```

### **Adjusting the spacing around subplots**

```python
from pickle import FALSE
# An example that shrink the spacing of the subplots to zero
# change the value of wspace and hspace
# sharex=True parameter means that the subplots will share the same x-axis
# The "fig" variable is a reference to the entire figure
# The "axes" is a 2D array of subplot references
fig, axes = plt.subplots(2, 2, sharex = True, sharey = False)

# The outer loop (for i in range(2)) iterates over rows
# The inner loop (for j in range(2)) iterates over columns
# Top left -> Top right -> bottom left -> bottom right
#(0, 0) = topleft
#(0, 1) = top right
#(1, 0) = bottom left
#(1, 1) = bottom right
for i in range(2): #0, 1 for i
  for j in range(2): # 0, 1 for j
    axes[i, j].hist(rng.standard_normal(500), bins = 50, color = "orange", alpha = .5)

#adjust space around subplot
# wspace = width of subplot
# hspace = height of padding
fig.subplots_adjust(wspace = 0.5, hspace = .5)
```

# **Adding shapes to a plot**

```python
# Adding shapes to a plot
fig, ax = plt.subplots() #alwys use code for creating subplots

# create a ractangle shape
# .4 = the width
# .2 = .75+ .2 = 9.5 = height
rect = plt.Rectangle((0.2, .75), .4, .2, color = "#483d8b", alpha = .3)

# create a circle
# .15 = radius of circle
circ = plt.Circle(( .7, .2), .15, color = '#556b2f', alpha = .3)

#create triagle shape
# hexagon = polygon of 6 sides
trig = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]], color = "#9370db", alpha = .5)

# all together
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(trig);
```

```python
# Adding shapes to a plot : Attention
import matplotlib.patches as mpatches
fig, ax = plt.subplots()

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color="black", alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color="blue", alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
                   color="green", alpha=0.5)

#Square shape
sqr = plt.Rectangle((0.3, 0.1), 0.2, 0.2, color="black", alpha=0.9)

# add ellipse
# (0.5, 0.5) -> This tuple specifies the center of the ellipse
# 0.2 -> width of the ellipse
# 0.1 -> height of the ellipse
ellipse = mpatches.Ellipse((.5, .5), 0.2, 0.1, color = 'orange')


# add hexagon
# 6 -> specify number of sides in the regular polygon
# 0.1 -> distance from the center of the circumcircle (the point at the center of the hexagon) to any of its vertices
hex = mpatches.RegularPolygon((.7,.8), 6, .1, color = 'red')

# add all subplots together
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
ax.add_patch(sqr)
ax.add_patch(ellipse)
ax.add_patch(hex);
```

```python
# Using the crosstab() function
# To compute a simple cross-tabulation of two (or more) factors.
# create two factors data frame
party_counts = pd.crosstab(tips['day'], tips['size'])
party_counts
```

```python
# Extract party_count from col 2 to 5
party_counts_edit = party_counts.loc[:, 2:5]
party_counts_edit
```

```python
# Normalize each row to sum to 1
# axis="index" argument indicates that the division should be done along the rows
# doing some data wrangling of the data frame
party_pcts = party_counts_edit.div(party_counts_edit.sum(axis = 'columns'),
                                   axis = "index")
party_pcts


```
